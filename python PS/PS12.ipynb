### Perform the following operations using Python on census bureau databset(Adult data sets).
### a.Create data subsets for different Country, Sex, race.
### b. Merge two subsets
### c. Sort Data using customer ratings.

### f. Casting data to wide format

```python
import pandas as pd
```

```python
df = pd.read_csv("adult_dataset.csv")
df.head()
```

###  Create data subsets for type of post

```python
# Subsets based on Country
india_df = df[df['native-country'] == 'India']
us_df = df[df['native-country'] == 'United-States']

# Subsets based on Sex
male_df = df[df['gender'] == 'Male']
female_df = df[df['gender'] == 'Female']

# Subsets based on Race
white_df = df[df['race'] == 'White']
black_df = df[df['race'] == 'Black']
```

```python
print("Subset - India:", india_df.shape)
india_df.head()
```

```python
print("Subset - Male:", male_df.shape)
male_df.head()
```

```python
print("Subset - White Race:", white_df.shape)
white_df.head()
```

### Merge two subsets

```python
merged_sex_df = pd.concat([male_df, female_df])
print("Merged Sex Data Shape:", merged_sex_df.shape)
merged_sex_df.head()
```

### Sort Data using customer ratings.

```python
sorted_df = df.sort_values(by='hours-per-week', ascending=False)
sorted_df.head()
```

### Transposing Data

```python
# Transposing means converting rows to columns and columns to rows.

transposed_df = df.transpose()
print("Transposed Data:\n")
transposed_df.head()
```

### Melting Data to long format

```python
# Melting is the process of turning wide-format data (where each variable has its own column) into long-format data
# (where there are only two or three columns: one for the variable name and one for the value).

melted_df = pd.melt(
    df,
    id_vars=['gender', 'race'],  # Identifiers to retain
    value_vars=['age', 'educational-num', 'hours-per-week'],  # Columns to melt
    var_name='Attribute',
    value_name='Value'
)

print("🔹 Melted Data (Long Format):")
print(melted_df.head())
```

### Casting data to wide format

```python
# Casting involves transforming long-format data into a wide format, where each unique value in one column becomes its own separate column.

pivot_df = df.pivot_table(
    values='hours-per-week',
    index='race',
    columns='gender',
    aggfunc='mean'
)

print("🔹 Casted Data (Wide Format - Avg Hours/Week):")
print(pivot_df.head())
```
